```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib import colors

from itertools import cycle
```

```python
def to_point(line):
    try:
        mapping = ['number', 'x', 'y', 'demand', 'ready_time', 'due_date', 'service_time']
        line = line.replace('\n', '').split()
        point = {k:int(v) for k,v in zip(mapping, line)}
        return point
    except ValueError:
        raise ValueError("Can't read problem file")
```

```python
def convert_solution_to_graph(points, solution=None):
    G = nx.Graph()
    G.add_nodes_from(range(len(points)))
    if solution is not None:
        for route in solution:
            nx.add_path(G, list(map(int, route.split()[::2])))
    return G
```

```python
def get_paths_from_solution(solution):
    paths = []
    for route in solution:
        path = []
        route = list(map(int, route.split()[::2]))
        for i in range(len(route) - 1):
            path.append((route[i], route[i+1]))
        paths.append(path)
    return paths
```

```python
def draw_solution(problem_path, solution):
    
    with open(problem_path, 'r') as f:
        data = f.readlines()
        points = []
        for d in data[9:]:
            try:
                points.append(to_point(d))
            except ValueError:
                print("Warn: There is maybe an error in the problem file")
        pos = {point['number']: (point['x'], point['y']) for point in points}

    col = cycle(colors.TABLEAU_COLORS)
    # pos = nx.spring_layout(G)

    fig, ax = plt.subplots(figsize=(15,15), dpi=300)
    # ax.grid()

    G = convert_solution_to_graph(points, solution)
    paths = get_paths_from_solution(solution)
    
#     nx.draw_networkx_labels(G, pos=pos, labels=labels, ax=ax)
    for path in paths:
        # print(path)
        path_color = next(col)
        nodes_in_path = {x for p in path for x in p}
        node_sizes = [point['demand'] for point in points if point['number'] in nodes_in_path]

        labels = {filter(lambda x: x['number'] in nodes_in_path, points)}
#         G = nx.DiGraph(G)
        nx.draw_networkx_edges(G, pos=pos, edgelist=path, ax=ax, edge_color=path_color)
        nx.draw_networkx_nodes(G, pos=pos, nodelist=nodes_in_path, node_size=node_sizes, node_color=path_color, ax=ax)
    plt.savefig(f"./solutions/{problem_path.split('.')[-2].split('/')[-1]}.graph.png", dpi=300)
```
